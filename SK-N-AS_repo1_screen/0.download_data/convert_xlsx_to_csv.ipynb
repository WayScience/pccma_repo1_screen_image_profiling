# Convert original Excel platemap file to individual CSVs per plate

```python
import pandas as pd
from pathlib import Path
```

```python
# Define paths
xlsx_folder = Path("orig_xlsx_files")
metadata_folder = Path("metadata")
metadata_folder.mkdir(exist_ok=True)  # Ensure metadata folder exists

# Initialize a list to store barcode and file mapping
barcode_platemap = []

# Iterate through all xlsx files in the folder
for xlsx_file in xlsx_folder.glob("*.xlsx"):
    # Read the Excel file
    df = pd.read_excel(xlsx_file)

    # Identify unique plates in 'Plate Barcode' column
    unique_plates = df["Plate Barcode"].unique()
    print(f"Processing {xlsx_file.name}: Found {len(unique_plates)} unique plates.")

    # Split into individual CSV files per unique plate
    for plate in unique_plates:
        plate_df = df[df["Plate Barcode"] == plate]
        output_file = metadata_folder / f"{plate}_platemap.csv"
        plate_df.to_csv(output_file, index=False)
        print(f"Saved {output_file}")

        # Append barcode and file name to the list
        barcode_platemap.append({"Plate Barcode": plate, "File Name": output_file.name})
```

```python
# Create a DataFrame for barcode and file mapping
barcode_platemap_df = pd.DataFrame(barcode_platemap)
barcode_platemap_file = metadata_folder / "barcode_platemap.csv"
barcode_platemap_df.to_csv(barcode_platemap_file, index=False)
print(f"Saved barcode mapping file: {barcode_platemap_file}")
```
